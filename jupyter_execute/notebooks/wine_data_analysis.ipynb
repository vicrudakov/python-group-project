# Wine Data Analysis

```python
import os

from wine_data_analysis.download_wine_data import DownloadWineData
from wine_data_analysis.visualize_wine_data import VisualizeWineData
from wine_data_analysis.classify_wine_data import ClassifyWineData
```

## Set local variables

```python
data_download_path = os.path.abspath("")
data_file_name = 'wine_data.csv'
```

## Download data

```python
data_download = DownloadWineData(data_download_path, data_file_name)
data_download.set_cwd()
data_download.download_data()
df = data_download.load_data()
```

## Plot data

```python
data_visualize = VisualizeWineData(df)
```

### Draw histograms

```python
data_visualize.draw_histograms();
```

### Draw boxplots

```python
data_visualize.draw_boxplots();
```

## Use classification methods

```python
data_classify = ClassifyWineData(df, 89)
y_pred_logistic_regression = data_classify.logistic_regression()
y_pred_decision_tree = data_classify.decision_tree()
y_pred_knn = data_classify.knn()
```

### Logistic regression classification metrics

```python
print(f'Accuracy: {data_classify.accuracy(y_pred_logistic_regression):.3f}')
print(f'Average precision: {data_classify.average_precision(y_pred_logistic_regression):.3f}')
print(f'Average recall: {data_classify.average_recall(y_pred_logistic_regression):.3f}')
```

### Decision tree classification metrics

```python
print(f'Accuracy: {data_classify.accuracy(y_pred_decision_tree):.3f}')
print(f'Average precision: {data_classify.average_precision(y_pred_decision_tree):.3f}')
print(f'Average recall: {data_classify.average_recall(y_pred_decision_tree):.3f}')
```

### KNN classification metrics

```python
print(f'Accuracy: {data_classify.accuracy(y_pred_knn):.3f}')
print(f'Average precision: {data_classify.average_precision(y_pred_knn):.3f}')
print(f'Average recall: {data_classify.average_recall(y_pred_knn):.3f}')
```
